# Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots inside notebook
%matplotlib inline

# Set style
sns.set(style="whitegrid")
```

# Load the Dataset

```python
# Load dataset
df = pd.read_csv("sample_data/california_housing_train.csv")  # Update file name if needed

# Display first 5 rows
print(df.head())

# Dataset info
print(df.info())

# Statistical summary
print(df.describe())
```

# Survival Count Plot

```python
plt.figure(figsize=(10,6))
sns.countplot(x='housing_median_age', data=df)
plt.title("Distribution of Housing Median Age")
plt.xlabel("Housing Median Age")
plt.ylabel("Count")
plt.show()
```

# Survival By Gender

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='longitude', y='latitude', hue='median_house_value', data=df, palette='viridis', s=8)
plt.title("Housing Prices by Location (Median House Value)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

# Survival By Passenger Class

```python
plt.figure(figsize=(10,6))
sns.countplot(x='housing_median_age', data=df)
plt.title("Distribution of Housing Median Age")
plt.xlabel("Housing Median Age")
plt.ylabel("Count")
plt.show()
```

# Age Distributions

```python
plt.figure(figsize=(10,6))
sns.histplot(df['housing_median_age'], bins=30, kde=True)
plt.title("Distribution of Housing Median Age")
plt.xlabel("Housing Median Age")
plt.ylabel("Frequency")
plt.show()
```

# Survival By Age

```python
plt.figure(figsize=(8,4))
sns.boxplot(y='median_house_value', data=df)
plt.title("Distribution of Median House Value")
plt.ylabel("Median House Value")
plt.show()
```

# Fare Distribution By Class

```python
plt.figure(figsize=(8,4))
sns.boxplot(y='median_income', data=df)
plt.title("Distribution of Median Income")
plt.ylabel("Median Income")
plt.show()
```

## Correlation Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
