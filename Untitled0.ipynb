```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

num = int(input("Enter a non-negative number: "))

if num < 0:
    print("Factorial is not defined for negative numbers.")
elif num == 0:
    print("The factorial of 0 is 1.")
else:
    print(f"The factorial of {num} is {factorial(num)}.")
```

# Task
Develop a simple expert system in Python that uses a rule-based approach to make decisions or recommendations within a clearly defined domain (e.g., recommending a drink based on weather/mood, or a simple diagnostic system). Implement a rule base using 'if-then' statements, an inference engine to process user input and apply rules, and a command-line interface for interaction. Finally, test the system and provide a summary of its domain and decision-making process.

## Define Expert System Topic

### Subtask:
Clearly define a specific, simple domain for the expert system (e.g., recommending a type of drink based on weather/mood, or troubleshooting a simple device). This will help in creating relevant rules.


### Expert System Domain Definition

**Chosen Domain:** Recommending a type of drink based on weather and user preference.

**Input Factors:**
1.  **Weather:** (e.g., 'hot', 'cold', 'mild')
2.  **User Preference:** (e.g., 'coffee', 'tea', 'juice', 'soda')

**Possible Outputs (Drink Recommendations):**
*   'Iced Coffee'
*   'Hot Coffee'
*   'Iced Tea'
*   'Hot Tea'
*   'Orange Juice'
*   'Lemonade'
*   'Cola'
*   'Water'

**General Logic/Decision-Making Process:**
The system will take the current weather condition and the user's general drink preference as input. Based on these two factors, it will apply a set of rules to suggest an appropriate drink. For example, if it's hot and the user prefers coffee, it might recommend 'Iced Coffee'. If it's cold and they prefer tea, 'Hot Tea' would be suggested. There will also be a default or general recommendation like 'Water' if no specific preference or a neutral preference is given.

## Design Rule Base

### Subtask:
Develop a set of 'if-then' rules based on the chosen topic (drink recommendation based on weather and user preference). These rules will form the core logic of the expert system and should be represented in a structured way.


**Reasoning**:
The subtask requires developing a set of 'if-then' rules for drink recommendations based on weather and user preference, stored in a Python data structure. I will create a list of dictionaries, where each dictionary represents a rule with 'weather', 'preference', and 'recommendation'.



```python
rules = [
    # Hot Weather Recommendations
    {"weather": "hot", "preference": "coffee", "recommendation": "Iced Coffee"},
    {"weather": "hot", "preference": "tea", "recommendation": "Iced Tea"},
    {"weather": "hot", "preference": "juice", "recommendation": "Fresh Orange Juice"},
    {"weather": "hot", "preference": "soda", "recommendation": "Lemon-Lime Soda"},

    # Cold Weather Recommendations
    {"weather": "cold", "preference": "coffee", "recommendation": "Hot Coffee"},
    {"weather": "cold", "preference": "tea", "recommendation": "Hot Herbal Tea"},
    {"weather": "cold", "preference": "juice", "recommendation": "Hot Apple Cider"},

    # Mild Weather Recommendations
    {"weather": "mild", "preference": "coffee", "recommendation": "Latte"},
    {"weather": "mild", "preference": "tea", "recommendation": "Green Tea"},
    {"weather": "mild", "preference": "juice", "recommendation": "Cranberry Juice"},

    # General Preference for Water
    {"weather": "hot", "preference": "water", "recommendation": "Chilled Mineral Water"},
    {"weather": "cold", "preference": "water", "recommendation": "Room Temperature Water"},
    {"weather": "mild", "preference": "water", "recommendation": "Sparkling Water"},

    # Default rule if specific preference not found
    {"weather": None, "preference": None, "recommendation": "Water"}
]

print(f"Defined {len(rules)} recommendation rules.")
# print the first 3 rules to show structure
for rule in rules[:3]:
    print(rule)
```

## Implement Inference Engine

### Subtask:
Write Python code that takes user input, matches it against the rule base, and infers a decision. This will involve querying the user for information and applying the rules sequentially or using a simple forward/backward chaining approach.


**Reasoning**:
To implement the inference engine, I will define a Python function `get_recommendation` that iterates through the predefined rules to find a match based on user input for weather and preference, including a fallback for the default rule.



```python
def get_recommendation(weather_input, preference_input):
    """
    Infers a drink recommendation based on weather and user preference.
    """
    for rule in rules:
        # Check for specific rule matches first
        if rule['weather'] is not None and rule['preference'] is not None:
            if rule['weather'].lower() == weather_input.lower() and \
               rule['preference'].lower() == preference_input.lower():
                return rule['recommendation']

    # Fallback to the default rule if no specific match is found
    for rule in rules:
        if rule['weather'] is None and rule['preference'] is None:
            return rule['recommendation']

    return "No recommendation found (this should not happen if a default rule exists)."

print("Inference engine function 'get_recommendation' has been defined.")
```

## Create User Interface

### Subtask:
Develop a simple command-line interface to allow users to interact with the expert system by answering questions and receiving recommendations or diagnoses.


**Reasoning**:
To develop the command-line interface, I will prompt the user for weather and preference inputs, use the previously defined `get_recommendation` function to obtain a recommendation, and then display the result.



```python
print("\n--- Drink Recommendation Expert System ---")
weather_input = input("What is the current weather? (hot, cold, mild): ")
preference_input = input("What is your preferred drink type? (coffee, tea, juice, soda, water): ")

recommendation = get_recommendation(weather_input, preference_input)

print(f"\nBased on your input, we recommend: {recommendation}")
print("----------------------------------------")
```

**Reasoning**:
The previous test for 'cold' weather and 'coffee' preference incorrectly returned 'Water'. This was due to a trailing space in the user's input for weather ('cold ') and a potential logical flaw in the default rule handling within the `get_recommendation` function. I need to update the `get_recommendation` function to strip whitespace from user inputs and ensure the default rule is only applied if no specific rule is matched.



```python
def get_recommendation(weather_input, preference_input):
    """
    Infers a drink recommendation based on weather and user preference.
    """
    # Strip whitespace from inputs to handle potential user errors
    weather_input_cleaned = weather_input.strip().lower()
    preference_input_cleaned = preference_input.strip().lower()

    # First, try to find a specific rule match
    for rule in rules:
        if rule['weather'] is not None and rule['preference'] is not None:
            if rule['weather'].lower() == weather_input_cleaned and \
               rule['preference'].lower() == preference_input_cleaned:
                return rule['recommendation']

    # If no specific match is found, find and return the default rule
    for rule in rules:
        if rule['weather'] is None and rule['preference'] is None:
            return rule['recommendation']

    return "No recommendation found (this should not happen if a default rule exists)."

print("Inference engine function 'get_recommendation' has been updated to handle whitespace and correct default rule logic.")
```

**Reasoning**:
The `get_recommendation` function has been updated to handle input more robustly. Now, it's crucial to re-test the system with the same inputs that previously failed (e.g., 'cold' and 'coffee') to verify the fix and ensure the system behaves as expected.



```python
print("\n--- Retesting Drink Recommendation Expert System after fix ---")
weather_input = input("What is the current weather? (hot, cold, mild): ")
preference_input = input("What is your preferred drink type? (coffee, tea, juice, soda, water): ")

recommendation = get_recommendation(weather_input, preference_input)

print(f"\nBased on your input, we recommend: {recommendation}")
print("----------------------------------------")
```
