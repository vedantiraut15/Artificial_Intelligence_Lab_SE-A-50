 Generate a normal distribution and plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data from normal distribution
# mean = 0, standard deviation = 1, sample size = 1000
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')

# Labels and title
plt.title("Histogram of Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")

# Show plot
plt.show()
```

 Compute mean, median, variance, and standard deviation for a dataset.

```python
import numpy as np

# Example dataset
data = [10, 20, 30, 40, 50]

# Calculations
mean = np.mean(data)
median = np.median(data)
variance = np.var(data)
std_dev = np.std(data)

print("Dataset:", data)
print("Mean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

Simulate a coin toss experiment and compute probabilities.

```python
import random

# Number of tosses
n = 1000

# Simulate coin tosses
results = [random.choice(['H', 'T']) for _ in range(n)]

# Count outcomes
heads = results.count('H')
tails = results.count('T')

# Compute probabilities
p_heads = heads / n
p_tails = tails / n

print("Total tosses:", n)
print("Heads:", heads)
print("Tails:", tails)
print("Probability of Heads:", p_heads)
print("Probability of Tails:", p_tails)
```
